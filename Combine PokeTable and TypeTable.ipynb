```python
# Import Dependencies #
import pandas as pd
import requests
from sqlalchemy import create_engine
```

```python
# URL to scrape #
url = "https://pokemondb.net/type"
```

```python
# Fancy code to avoid 403: Forbidden error #
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
r = requests.get(url, headers=header)

# Use Pandas read_html function to pull table and rename columns #
tables = pd.read_html(r.text)
type_eff_df = tables[0]
type_eff_df.columns = ["Type", "Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", 
                      "Flying", "Psychic", "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"]
type_eff_df
```

```python
# Fill NaN cells with 1 and set index as "Type" #
type_eff_df.fillna(1, inplace=True)
type_eff_df = type_eff_df.set_index("Type")
type_eff_df.replace("½", .5, inplace=True)
```

```python
# Replace "1/2" character with integer .5 for maths later and transpose to create a weakness table instead of attack table #
type_eff_df.replace("½", .5, inplace=True)
type_weak_df = type_eff_df.T
type_weak_df = type_weak_df.apply(pd.to_numeric)
type_weak_df
```

```python
# Read in csv and convert to Pandas df for some transforming! #
abil_type_df = pd.read_csv("Type Effectiveness by Ability.csv")
abil_type_df = abil_type_df.set_index("Unnamed: 0")
abil_type_df
```

```python
# Read in csv and convert to Pandas df for some transforming! #
poke_type_df = pd.read_csv("Output/pokedex_w_ability.csv")
poke_type_df = poke_type_df.iloc[:,1:]
poke_type_df.head()
```

```python
# Create empty list and dictionary to store all the things #
effs = []
dictionary = {};

# Create list of reductive abilities #
red_list = ["Solid Rock", "Filter", "Prism Armor"]

# Create types list that will be used to iterate through column names in for loop below #
types = ["Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", "Flying", "Psychic", 
         "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"]

# use iterrows() to iterate through all rows in Pandas data frame #
for (index, row) in poke_type_df.iterrows():
   
    # set type 1 and row containing all type1 effectiveness values equal to variables #
    type1 = row[2]
    eff1_row = type_weak_df[type_weak_df.index == type1]
    
    # iterate through types/columns in current row #
    for type in types:
        # Find effectiveness value for current type in row#
        eff1 = eff1_row[type][0]
        
        # If type2 contains a value, do same as above for type2 and multiply effectiveness values for total #
        if pd.notnull(row[3]):
            type2 = row[3]
            eff2_row = type_weak_df[type_weak_df.index == type2]
            eff2 = eff2_row[type][0]
        else:
            eff2 = 1
        
        # include defensive ability in calculation #
        if pd.notnull(row[4]) and row[4] not in red_list:
            abil = row[4]
            eff_abil_row = abil_type_df[abil_type_df.index == abil]
            eff_abil = eff_abil_row[type][0]
        else:
            eff_abil = 1
            
        eff = eff1*eff2*eff_abil
        
        # include reductive abilities in calculation #
        if row[4] in red_list:
            eff= eff * .75
        
        # append all total effectiveness values to effs list #
        effs.append(eff)
        
    # add effectiveness and 
    dictionary[index] = effs
    effs = []
        
    
weaknesses_df = pd.DataFrame(dictionary).T
weaknesses_df.columns = types
weaknesses_df
```

```python
# Combine data frame using concatenate into its final form #
to_concat = [poke_type_df, weaknesses_df]
combined_df = pd.concat(to_concat, axis=1)
combined_df = combined_df.rename(index=str, columns={"name":"pokemon"})
```

```python
# Replace Nidoran special gender characters with strings #
combined_df.replace("Nidoran♂", "Nidoran(M)", inplace=True)
combined_df.replace("Nidoran♀", "Nidoran(M)", inplace=True)
combined_df.head()
```

```python
# Save to csv #
combined_df.to_csv("Output/Final_Table.csv")
```
