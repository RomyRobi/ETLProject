```python
# Import Dependencies #
import pandas as pd
import requests
```

```python
# URL to scrape #
url = "https://pokemondb.net/type"
```

```python
# Fancy code to avoid 403: Forbidden error #
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
r = requests.get(url, headers=header)

# Use Pandas read_html function to pull table and rename columns #
tables = pd.read_html(r.text)
type_eff_df = tables[0]
type_eff_df.columns = ["Type", "Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", 
                      "Flying", "Psychic", "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"]
type_eff_df
```

```python
# Fill NaN cells with 1 and set index as "Type" #
type_eff_df.fillna(1, inplace=True)
type_eff_df = type_eff_df.set_index("Type")
```

```python
# Replace "1/2" character with integer .5 for maths later and transpose to create a weakness table instead of attack table #
type_eff_df.replace("½", .5, inplace=True)
type_weak_df = type_eff_df.T
type_weak_df.columns = ["Normal", "Fire", "Water", "Electric", "Grass", "Ice", "Fighting", "Poison", "Ground", 
                      "Flying", "Psychic", "Bug", "Rock", "Ghost", "Dragon", "Dark", "Steel", "Fairy"]
type_weak_df = type_weak_df.iloc[1:,:]
type_weak_df = type_weak_df.apply(pd.to_numeric)
type_weak_df
```
